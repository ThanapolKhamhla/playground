# Euler's Disk

[Euler’s disk](https://en.wikipedia.org/wiki/Euler%27s_Disk) is the rigid–body motion of a slightly tilted circular disk that spins and precesses on a horizontal surface, emitting a rising chirp before stopping abruptly. The problem unites nonholonomic rolling kinematics, gravity, and dissipation.

## Geometry, frames, inertia
- Disk: mass $m$, radius $a$. Principal moments at the centre $O$:
$$
I_{11}=I_{22}=kma^2\qquad I_{33}=2kma^2
$$
- $\hat{\mathbf z}$: vertical unit vector (up).
$\hat{\mathbf n}$: disk symmetry axis. Tilt $\alpha$ defined by $\hat{\mathbf z}\cdot\hat{\mathbf n}=\cos\alpha$.
- Kinematic rates:
	- $\Omega$: precession about $\hat{\mathbf z}$,
	- $\omega_{\rm rel}$: spin about $\hat{\mathbf n}$ relative to the precessing frame.

With constant tilt ($\dot\alpha=0$), write the angular velocity as
$$
\boldsymbol{\omega}=\Omega\hat{\mathbf z}+\omega_{\rm rel}\hat{\mathbf n}. \tag{1}
$$

## Contact geometry and no-slip

Let $C$ be the contact point with the plane. The vector from the centre to the contact is the projection of $-\hat{\mathbf z}$ into the disk plane, scaled to length $a$:
$$
\boldsymbol{\rho}\equiv \overrightarrow{OC}
= -\frac{a}{\sin\alpha}\big(\hat{\mathbf z}-\cos\alpha\hat{\mathbf n}\big). \tag{2}
$$

No-slip at $C$ requires the contact point velocity to vanish:
$$
\mathbf v_C=\mathbf v_O+\boldsymbol{\omega}\times\boldsymbol{\rho}=\mathbf 0. \tag{3}
$$

In the observed steady solution the centre is instantaneously at rest ($\mathbf v_O=\mathbf 0$). Then
$$
\boldsymbol{\omega}\times\boldsymbol{\rho}=\mathbf 0
\quad\Longrightarrow\quad
\boldsymbol{\omega}\ \parallel\ \boldsymbol{\rho}. \tag{4}
$$

Comparing coefficients of $\hat{\mathbf z}$ and $\hat{\mathbf n}$ in (1) and (2) gives
$$
\boxed{\omega_{\rm rel}=-\Omega\cos\alpha}. \tag{5}
$$

Choose body axes with $\hat{\mathbf e}_3=\hat{\mathbf n}$ and $\hat{\mathbf e}_1$ along the projection of $\hat{\mathbf z}$ into the disk plane. Using (1)–(5),
$$
\omega_1=\Omega\sin\alpha\qquad \omega_2=0\qquad \omega_3=\boldsymbol{\omega}\cdot\hat{\mathbf n}=0. \tag{6}
$$
(Thus the lab-frame spin about the disk axis vanishes in this ideal steady state, explaining the apparent “face freeze.”)


## Dynamics about the contact point

Take moments about $C$. The contact forces act at $C$, so they exert no torque about $C$. Only gravity contributes:
$$
\boldsymbol{\tau}_C=\boldsymbol{\rho}\times(-mg\hat{\mathbf z})\qquad
|\boldsymbol{\tau}_C|=mga\sin\alpha. \tag{7}
$$

Angular momentum about $C$ satisfies
$$
\dot{\mathbf H}_C=\boldsymbol{\tau}_C. \tag{8}
$$

Because the centre is instantaneously at rest in the steady motion, the angular momentum about $C$ equals that about $O$:
$$
\mathbf H_C=\mathbf H_O=\mathbf I_O\boldsymbol{\omega}. \tag{9}
$$

With (6) and $\mathbf I_O=\mathrm{diag}(kma^2,kma^2,2kma^2)$,
$$
\mathbf H_C = kma^2\omega_1\hat{\mathbf e}_1
= kma^2\Omega\sin\alpha\hat{\mathbf e}_1,
\qquad
|\mathbf H_C| = kma^2\Omega\sin\alpha. \tag{10}
$$

In the steady motion, $\mathbf H_C$ precesses about $\hat{\mathbf z}$ at rate $\Omega$; the angle between $\mathbf H_C$ and $\hat{\mathbf z}$ is $\alpha$, so
$$
|\dot{\mathbf H}_C|=\Omega,|\mathbf H_C|\sin\alpha
=\Omega(kma^2\Omega\sin\alpha)\sin\alpha. \tag{11}
$$

Equate (11) with the torque magnitude (7) yeild:
$$
\Omega(kma^2\Omega\sin\alpha)\sin\alpha = mga\sin\alpha
\Longrightarrow
\boxed{\Omega^{2}=\frac{g}{ak\sin\alpha}}. \tag{12}
$$


## Immediate corollaries
- From (5): $\omega_{\rm rel}=-\Omega\cos\alpha$; the apparent face spin is
$$
\Omega-\omega_{\rm rel}=\Omega(1-\cos\alpha)\xrightarrow[\alpha\to0]{}0,
$$
while $\Omega\sim(\sin\alpha)^{-1/2}\to\infty$ — the audible chirp.
- Kinetic and potential energies (with $\dot\alpha=0$ and (6)):
$$
T=\tfrac12\boldsymbol{\omega}^{\top}\mathbf I_O\boldsymbol{\omega}
= \tfrac12kma^2\Omega^2\sin^2\alpha\qquad
V=mga\sin\alpha.
$$
Eliminating $\Omega$ via (12):
$$
\boxed{E=T+V=\tfrac{3}{2}m g a \sin\alpha}. \tag{13}
$$


## Power-law relations near the stop (finite-time singularity)

Let $t_f$ be the stop time, $\alpha(t)$ the tilt (to the vertical), $\Omega(t)$ the precession rate, and $E(t)$ the mechanical energy. In the late stage (small $\alpha$) the steady relations give
$$
E = \tfrac{3}{2}m g a\alpha
\qquad
\Omega^2 = \frac{g}{a k\alpha}
\ \ \Rightarrow\ 
\Omega \propto \alpha^{-1/2}.
\tag{14}
$$

To get the power laws in time, model dissipation by a rate
$$
P_{\rm diss} \equiv -\frac{dE}{dt} = C\Omega^{p}\alpha^{q},
\tag{15}
$$
which covers the common mechanisms (rolling/contact losses, air drag variants, etc.). Using $E\propto \alpha$ and (14),
$$
\frac{d\alpha}{dt} = -K\alpha^{-p/2+q} \equiv -K\alpha^{\beta}.
\tag{16}
$$
Solving gives
$$
\boxed{
\alpha(t)\ \propto\ (t_f-t)^{\frac{1}{1-\beta}}\qquad
\Omega(t)\ \propto\ (t_f-t)^{-\frac{1}{2(1-\beta)}}\qquad
E(t)\ \propto\ (t_f-t)^{\frac{1}{1-\beta}}
}
\quad\big(\beta=-\tfrac{p}{2}+q\big).
\tag{17}
$$

## Rolling/contact friction 

High-speed measurements identify rolling/contact losses as the primary mechanism. e.g. [Easwar–Rouyer–Menon](https://journals.aps.org/pre/abstract/10.1103/PhysRevE.66.045102).

Rolling-resistance torque $M_r\sim\mu_rN a\approx\mu_r m g a$; power $P=M_r\Omega \Rightarrow p=1,q=0$. 

Then $\beta=-\tfrac12$ and
$$
\boxed{\ \alpha\propto (t_f-t)^{2/3}\quad \Omega\propto (t_f-t)^{-1/3}\quad E\propto (t_f-t)^{2/3}\ }.
\tag{18}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from dataclasses import dataclass

g = 9.81

@dataclass
class Disk:
    m: float = 0.20
    a: float = 0.03
    k: float = 0.25  # uniform thin disk

def Omega_precession(alpha, disk: Disk):
    return np.sqrt(g / (disk.a * disk.k * np.sin(alpha)))

def omega_rel(alpha, disk: Disk):
    return Omega_precession(alpha, disk) * np.cos(alpha)

def omega_face(alpha, disk: Disk):
    Ω = Omega_precession(alpha, disk)
    return Ω * (1 - np.cos(alpha))

def energy_steady(alpha, disk: Disk):
    return 1.5 * disk.m * g * disk.a * np.sin(alpha)
```

```python
disk = Disk()
alphas = np.deg2rad(np.linspace(0.5, 30.0, 400))

Ω = Omega_precession(alphas, disk)
ωr = omega_rel(alphas, disk)
ωf = omega_face(alphas, disk)
E = energy_steady(alphas, disk)

fig, axes = plt.subplots(1, 3, figsize=(10,3))

ax = axes[0]
ax.plot(np.rad2deg(alphas), Ω/(2*np.pi))
ax.set_xlabel("Tilt angle α (degrees)")
ax.set_ylabel("Ω/2π (Hz)")
ax.set_title("Precession chirp vs α")

ax = axes[1]
ax.plot(np.rad2deg(alphas), ωf/(2*np.pi))
ax.set_xlabel("Tilt angle α (degrees)")
ax.set_ylabel("(Ω-ω_rel)/2π (Hz)")
ax.set_title("Apparent face spin vs α")

ax = axes[2]
ax.plot(np.rad2deg(alphas), E)
ax.set_xlabel("Tilt angle α (degrees)")
ax.set_ylabel("E (J)")
ax.set_title("Energy ≈ (3/2) m g a sin α")

plt.tight_layout()
plt.show()
```

```python
t_f = 10.0
t = np.linspace(0, t_f-1e-3, 800)
alpha0 = np.deg2rad(20.0)
C = alpha0 / (t_f**(2/3))

alpha_t = C * (t_f - t)**(2/3)
Ω_t = Omega_precession(alpha_t, disk)
E_t = energy_steady(alpha_t, disk)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10,3))

ax = axes[0]
ax.plot(t, np.rad2deg(alpha_t)); 
ax.set_xlabel(r"$t (s)$"); 
ax.set_ylabel(r"$\alpha$ (deg)")
ax.set_title(r"$\alpha(t) \propto (t_f - t)^{2/3}$")


ax2 = axes[1]
ax2.plot(t, Ω_t/(2*np.pi)); 
ax2.set_xlabel("t (s)"); 
ax2.set_ylabel(r"$\Omega/2\pi (Hz)$")
ax2.set_title(r"$\Omega(t) \propto (t_f - t)^{-1/3}$")

ax3 = axes[2]
ax3.plot(t, E_t); 
ax3.set_xlabel(r"$t (s)$"); 
ax3.set_ylabel(r"$E (J)$")
ax3.set_title(r"$E(t) \propto (t_f - t)^{2/3}$"); 

fig.tight_layout()
plt.show()
```

## Viscous air drag — fully developed gap flow

Laminar shear across the gap of thickness $h\sim a\alpha$ gives roughly $P\propto\mu\Omega^{2}\alpha^{-1}$ ($p=2,q=-1$). Then $\beta=-2$ and
$$
\boxed{\ \alpha\propto (t_f-t)^{1/3}\quad \Omega\propto (t_f-t)^{-1/6}\quad E\propto (t_f-t)^{1/3}\ }.
\tag{19}
$$
as analysed by [Moffatt](https://arxiv.org/abs/gr-qc/0102046). This yields a shallower chirp than (5). 

```python
disk = Disk()
alphas = np.deg2rad(np.linspace(0.5, 30.0, 400))

Ω = Omega_precession(alphas, disk)
ωr = omega_rel(alphas, disk)
ωf = omega_face(alphas, disk)
E = energy_steady(alphas, disk)


t_f = 10.0
t = np.linspace(0, t_f-1e-3, 800)
alpha0 = np.deg2rad(20.0)
C = alpha0 / (t_f**(1/3))

alpha_t = C * (t_f - t)**(1/3)
Ω_t = Omega_precession(alpha_t, disk)
E_t = energy_steady(alpha_t, disk)

fig, axes = plt.subplots(1, 3, figsize=(10,3))

ax = axes[0]
ax.plot(t, np.rad2deg(alpha_t)); 
ax.set_xlabel(r"$t (s)$"); 
ax.set_ylabel(r"$\alpha$ (deg)")
ax.set_title(r"$\alpha(t) \propto (t_f - t)^{1/3}$")


ax2 = axes[1]
ax2.plot(t, Ω_t/(2*np.pi)); 
ax2.set_xlabel("t (s)"); 
ax2.set_ylabel(r"$\Omega/2\pi (Hz)$")
ax2.set_title(r"$\Omega(t) \propto (t_f - t)^{-1/6}$")

ax3 = axes[2]
ax3.plot(t, E_t); 
ax3.set_xlabel(r"$t (s)$"); 
ax3.set_ylabel(r"$E (J)$")
ax3.set_title(r"$E(t) \propto (t_f - t)^{1/3}$"); 

fig.tight_layout()
plt.show()
```

## Viscous air drag — oscillatory boundary layers 

Including finite-thickness boundary layers gives $P\propto \Omega^{3/2}\alpha^{-1/2}$ ($p=\tfrac32, q=-\tfrac12$). Then $\beta=-\tfrac54$ and
$$
\boxed{\ \alpha\propto (t_f-t)^{4/9}\quad \Omega\propto (t_f-t)^{-2/9}\quad E\propto (t_f-t)^{4/9}\ }.
\tag{20}
$$
See [Bildsten’s note](https://ar5iv.labs.arxiv.org/html/gr-qc/0102046)

```python
disk = Disk()
alphas = np.deg2rad(np.linspace(0.5, 30.0, 400))

Ω = Omega_precession(alphas, disk)
ωr = omega_rel(alphas, disk)
ωf = omega_face(alphas, disk)
E = energy_steady(alphas, disk)


t_f = 10.0
t = np.linspace(0, t_f-1e-3, 800)
alpha0 = np.deg2rad(20.0)
C = alpha0 / (t_f**(4/9))

alpha_t = C * (t_f - t)**(4/9)
Ω_t = Omega_precession(alpha_t, disk)
E_t = energy_steady(alpha_t, disk)

fig, axes = plt.subplots(1, 3, figsize=(10,3))

ax = axes[0]
ax.plot(t, np.rad2deg(alpha_t)); 
ax.set_xlabel(r"$t (s)$"); 
ax.set_ylabel(r"$\alpha$ (deg)")
ax.set_title(r"$\alpha(t) \propto (t_f - t)^{4/9}$")


ax2 = axes[1]
ax2.plot(t, Ω_t/(2*np.pi)); 
ax2.set_xlabel("t (s)"); 
ax2.set_ylabel(r"$\Omega/2\pi (Hz)$")
ax2.set_title(r"$\Omega(t) \propto (t_f - t)^{-2/9}$")

ax3 = axes[2]
ax3.plot(t, E_t); 
ax3.set_xlabel(r"$t (s)$"); 
ax3.set_ylabel(r"$E (J)$")
ax3.set_title(r"$E(t) \propto (t_f - t)^{4/9}$"); 

fig.tight_layout()
plt.show()
```
