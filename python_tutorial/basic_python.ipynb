## Example: Random Flight Polymer

Random flight polymer: assumes that the conformation of the polymer is entirely random. We can simulate a random flight polymer through a random walk by making each subsequent segment of polymer extend in a random direction and distance.

- start the polymer chain at coordinate (0, 0)
- define number of segments to add
- for each new segments, add a random value (-1,1) to the x-coordinate of the previous polymer end 
- add another random value to the y-coordinate. 
- each new coordinate is appended to a list of coordinates.

Common Random Functions in Python (`random` module)

| Function | Description | Example | Possible Output | Return Type |
|:----------|:-------------|:---------|:----------------|:-------------|
| `random()` | Returns a float uniformly distributed in **[0.0, 1.0)**. | `random()` | `0.3721` | `float` |
| `uniform(a, b)` | Returns a float uniformly distributed in **[a, b]**. | `uniform(-10, 5)` | `-3.14` | `float` |
| `randint(a, b)` | Returns a random **integer N** in **[a, b]** (inclusive). | `randint(1, 6)` | `4` | `int` |
| `gauss(mu, sigma)` | Returns a random float from a **Normal (Gaussian)** distribution. | `gauss(0, 1)` | `-0.56` | `float` |
| `seed(a=None)` | Initializes the random number generator (for reproducibility). | `seed(42)` | – | `None` |


```python
import random

# coordinate of first segment
x_coordinate = [0]
y_coordinate = [0]

segments = 1000 # number of polymer segments

for segment in range(segments):
    x_new = x_coordinate[segment] + random.uniform(-1, 1)
    y_new = y_coordinate[segment] + random.uniform(-1, 1)
    
    x_coordinate.append(x_new)
    y_coordinate.append(y_new)
```

```python
import matplotlib.pyplot as plt

plt.plot(x_coordinate, y_coordinate)
plt.show()
```
